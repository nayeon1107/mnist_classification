```python
import pandas as pd
import matplotlib.pyplot as plt

PATH = "/home/user/Desktop/na/mnist-classification/"
epochs = range(1,31)


lenet_base = pd.read_pickle(PATH+"LeNet5_base.pickle")
mlp_base = pd.read_pickle(PATH+"MLP_base.pickle")
lenet_drop = pd.read_pickle(PATH+"LeNet5_Dropout.pickle")
lenet_batchnorm = pd.read_pickle(PATH+"LeNet5_batchNorm.pickle")

train_acc_le = lenet_base['train_acc']
train_loss_le = lenet_base['train_loss']
test_acc_le = lenet_base['test_acc']
test_loss_le = lenet_base['test_loss']

train_acc_mlp = mlp_base['train_acc']
train_loss_mlp = mlp_base['train_loss']
test_acc_mlp = mlp_base['test_acc']
test_loss_mlp = mlp_base['test_loss']

train_acc_ledr = lenet_drop['train_acc']
train_loss_ledr = lenet_drop['train_loss']
test_acc_ledr = lenet_drop['test_acc']
test_loss_ledr = lenet_drop['test_loss']

train_acc_lebatch = lenet_batchnorm['train_acc']
train_loss_lebatch = lenet_batchnorm['train_loss']
test_acc_lebatch = lenet_batchnorm['test_acc']
test_loss_lebatch = lenet_batchnorm['test_loss']
```

```python
plt.plot(epochs, train_acc_le, 'b', label='LeNet5')
plt.plot(epochs, train_acc_mlp, 'c', label='CustomMLP')
plt.title('LeNet5 and CustomMLP Train accuracy')
plt.legend()
# plt.savefig('/home/user/Desktop/na/mnist-classification/base train accuracy.png')

plt.figure()

plt.plot(epochs, train_loss_le, 'b', label='LeNet5')
plt.plot(epochs, train_loss_mlp, 'c', label='CustomMLP')
plt.title('LeNet5 and CustomMLP Train Loss ')
# plt.savefig('/home/user/Desktop/na/mnist-classification/base train loss.png')

plt.figure()

plt.plot(epochs, test_acc_le, 'b', label='LeNet5')
plt.plot(epochs, test_acc_mlp, 'c', label='CustomMLP')
plt.title('LeNet5 and CustomMLP Test accuracy')
# plt.savefig('/home/user/Desktop/na/mnist-classification/base test accuracy.png')

plt.figure()

plt.plot(epochs, train_loss_le, 'b', label='LeNet5')
plt.plot(epochs, train_loss_mlp, 'c', label='CustomMLP')
plt.title('LeNet5 and CustomMLP Test Loss')
# plt.savefig('/home/user/Desktop/na/mnist-classification/base test loss.png')

plt.show()
```

```python
plt.plot(epochs, train_acc_le, 'b', label='LeNet5')
plt.plot(epochs, train_acc_ledr, 'b--', label='LeNet5 + Dropout(0.5)')
plt.title('Train accuracy between Dropout')
plt.legend()
# plt.savefig('/home/user/Desktop/na/mnist-classification/base train accuracy.png')

plt.figure()

plt.plot(epochs, train_loss_le, 'b', label='LeNet5')
plt.plot(epochs, train_loss_ledr, 'b--', label='LeNet5 + Dropout(0.5)')
plt.title('Train loss between Dropout')
# plt.savefig('/home/user/Desktop/na/mnist-classification/base train loss.png')

plt.figure()

plt.plot(epochs, test_acc_le, 'b', label='LeNet5')
plt.plot(epochs, test_acc_ledr, 'b--', label='LeNet5 + Dropout(0.5)')
plt.title('Test accuracy between Dropout')
# plt.savefig('/home/user/Desktop/na/mnist-classification/base train accuracy.png')

plt.figure()

plt.plot(epochs, test_loss_le, 'b', label='LeNet5')
plt.plot(epochs, test_loss_ledr, 'b--', label='LeNet5 + Dropout(0.5)')
plt.title('Test loss between Dropout')
# plt.savefig('/home/user/Desktop/na/mnist-classification/base train loss.png')

plt.show()
```

```python
plt.plot(epochs, train_acc_le, 'b', label='LeNet5')
plt.plot(epochs, train_acc_lebatch, 'b--', label='LeNet5 + BatchNorm')
plt.title('Train accuracy between BatchNorm')
plt.legend()
# plt.savefig('/home/user/Desktop/na/mnist-classification/base train accuracy.png')

plt.figure()

plt.plot(epochs, train_loss_le, 'b', label='LeNet5')
plt.plot(epochs, train_loss_lebatch, 'b--', label='LeNet5 + BatchNorm')
plt.title('Train loss between BatchNorm')
# plt.savefig('/home/user/Desktop/na/mnist-classification/base train loss.png')

plt.figure()

plt.plot(epochs, test_acc_le, 'b', label='LeNet5')
plt.plot(epochs, test_acc_lebatch, 'b--', label='LeNet5 + BatchNorm')
plt.title('Test accuracy between BatchNorm')
# plt.savefig('/home/user/Desktop/na/mnist-classification/base train accuracy.png')

plt.figure()

plt.plot(epochs, test_loss_le, 'b', label='LeNet5')
plt.plot(epochs, test_loss_lebatch, 'b--', label='LeNet5 + BatchNorm')
plt.title('Test loss between BatchNorm')
# plt.savefig('/home/user/Desktop/na/mnist-classification/base train loss.png')

plt.show()
```
